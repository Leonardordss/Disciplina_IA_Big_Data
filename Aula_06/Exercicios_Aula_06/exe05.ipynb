```python
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree
from sklearn.metrics import accuracy_score, precision_score, confusion_matrix
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import LabelEncoder
```

```python
#criando a base de dados
data = pd.DataFrame({
    'idade': [2, 6, 1, 4, 3, 5],
    'horas desde a ultima manutencao':[50, 200, 20, 150, 100, 180],
    'historico de falhas': [0, 2, 0, 1, 3, 2],
    'classifi': ['necessita', 'necessita', 'necessita', 'nao necessita', 'necessita', 'nao necessita']
})
```

```python
# Preparando os dados (com múltiplas variáveis)
x = np.array(data[['idade', 'horas desde a ultima manutencao', 'historico de falhas']])
y = np.array(data['classifi'])
```

```python
# Dividindo os dados em treino e teste
x_treino, x_teste, y_treino, y_teste = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
# Treinando o modelo de árvore de decisão
model = DecisionTreeClassifier()
model.fit(x_treino, y_treino)
```

```python
# Fazendo previsões
prev = model.predict(x_teste)
```

```python
# Calculando a acurácia
accuracy = accuracy_score(y_teste, prev)
```

```python
# Matriz de confusão
matrix = confusion_matrix(y_teste, prev)
```

```python
print(f'Acurácia: {accuracy}')
print(f'Matriz de Confusão: \n{matrix}')
```

```python
x=np.array(data['idade'])
y=np.array(data['classifi'])
x_treino, x_teste, y_treino, y_teste = train_test_split(x.reshape(-1,1), y.reshape(-1,1),
                                                        test_size=0.5, random_state=42)
model = DecisionTreeClassifier()
model.fit(x_treino, y_treino)
prev = model.predict(x_teste)
acurracy = accuracy_score(y_teste, prev)
matrix = confusion_matrix(y_teste, prev)
print(acurracy, matrix)
fig = plt.figure(figsize=(10,8))
tree.plot_tree(model, feature_names=x.tolist(), class_names=y.tolist(), filled=True)
plt.show()
```
