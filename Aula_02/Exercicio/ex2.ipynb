```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r'C:\Users\Admin-PC\Desktop\Disciplina_IA_Big_Data\Aula_02\monitoramento_tempo.csv')
```

```python

df['data'] = pd.to_datetime(df['data'])
df['ano'] = df['data'].dt.year

plt.figure(figsize=(10, 6))
plt.plot(df['ano'], df['velocidade do vento'], marker='o', color='b', linestyle='-', linewidth=2)

plt.title("Velocidade do Vento x Ano", fontsize=15)
plt.xlabel("Ano", fontsize=14)
plt.ylabel("Velocidade do Vento (m/s)", fontsize=14)

plt.legend(["Velocidade do Vento"], loc='best')
plt.grid(True)
plt.show()
```
