```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r'C:\Users\Admin-PC\Desktop\Disciplina_IA_Big_Data\Aula_02\monitoramento_tempo.csv')
```

```python
# Converter a coluna 'data' para datetime
df['data'] = pd.to_datetime(df['data'])

# Filtrar o período de maio a agosto de 2016
df_maio_agosto = df[(df['data'] >= '2016-05-01') & (df['data'] <= '2016-08-31')]

# Filtrar o ano de 2016
df_2016 = df[df['data'].dt.year == 2016]


fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(10, 12), sharex=True)

# Gráfico maior para maio a agosto de 2016
ax1.plot(df_maio_agosto['data'], df_maio_agosto['temperatura'])
ax1.set_title("Temperatura de Maio a Agosto de 2016", fontsize=16)
ax1.set_ylabel("Temperatura (°C)", fontsize=14)
ax1.grid(True)

# Gráfico menor para todo o ano de 2016
ax2.plot(df_2016['data'], df_2016['temperatura'])
ax2.set_title("Temperatura em 2016", fontsize=16)
ax2.set_ylabel("Temperatura (°C)", fontsize=14)
ax2.set_xlabel("Data", fontsize=14)
ax2.grid(True)

ax2.xaxis.set_major_formatter(plt.FuncFormatter(lambda x, _: pd.to_datetime(x).strftime('%m-%Y')))
ax2.xaxis.set_major_locator(plt.MaxNLocator(12))  

plt.tight_layout()  
plt.show()
```
