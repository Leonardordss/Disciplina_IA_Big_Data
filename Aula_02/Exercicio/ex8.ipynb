```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r'C:\Users\Admin-PC\Desktop\Disciplina_IA_Big_Data\Aula_02\monitoramento_tempo.csv')
```

```python

plt.figure(figsize=(10, 6))
plt.boxplot(df['pressão'], vert=True, patch_artist=True)

plt.title("Boxplot da Pressão", fontsize=15)
plt.ylabel("Pressão (hPa)", fontsize=14)
plt.xticks([1], ['Pressão'])  # Renomear o eixo x

plt.grid(True)
plt.show()
```
