```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r'C:\Users\Admin-PC\Desktop\Disciplina_IA_Big_Data\Aula_02\monitoramento_tempo.csv')
```

```python
# Converter a coluna 'data' para datetime
df['data'] = pd.to_datetime(df['data'])

# Filtrar o período de maio a agosto de 2017
df_maio_agosto_2017 = df[(df['data'] >= '2017-05-01') & (df['data'] <= '2017-08-31')]


fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(10, 12), sharex=True)

# Gráfico maior para temperatura de maio a agosto de 2017
ax1.plot(df_maio_agosto_2017['data'], df_maio_agosto_2017['temperatura'])
ax1.set_title("Temperatura de Maio a Agosto de 2017", fontsize=16)
ax1.set_ylabel("Temperatura (°C)", fontsize=14)
ax1.grid(True)

# Gráfico menor para umidade de maio a agosto de 2017
ax2.plot(df_maio_agosto_2017['data'], df_maio_agosto_2017['humidade'])
ax2.set_title("Umidade de Maio a Agosto de 2017", fontsize=16)
ax2.set_ylabel("Umidade (%)", fontsize=14)
ax2.set_xlabel("Data", fontsize=14)
ax2.grid(True)


ax2.xaxis.set_major_formatter(plt.FuncFormatter(lambda x, _: pd.to_datetime(x).strftime('%m-%Y')))
ax2.xaxis.set_major_locator(plt.MaxNLocator(12))  

plt.tight_layout()  
plt.show()
```
