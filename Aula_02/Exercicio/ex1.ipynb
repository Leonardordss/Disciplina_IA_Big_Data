```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r'C:\Users\Admin-PC\Desktop\Disciplina_IA_Big_Data\Aula_02\monitoramento_tempo.csv')
```

```python
# Converter a coluna 'data' para datetime e extrair o ano
df['data'] = pd.to_datetime(df['data'])
df['ano'] = df['data'].dt.year

plt.figure(figsize=(10, 6))
plt.plot(df['ano'], df['humidade'])

plt.title("Umidade x Ano", fontsize=15)
plt.xlabel("Ano")
plt.ylabel("Umidade")

plt.legend(["Umidade"], loc='best')


plt.show()

```
