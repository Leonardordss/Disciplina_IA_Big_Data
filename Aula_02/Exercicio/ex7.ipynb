```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r'C:\Users\Admin-PC\Desktop\Disciplina_IA_Big_Data\Aula_02\monitoramento_tempo.csv')
```

```python
plt.figure(figsize=(10, 6))
plt.boxplot(df['direção do vento'], vert=True, patch_artist=True)

plt.title("Boxplot da Direção do Vento", fontsize=15)
plt.ylabel("Direção do Vento (graus)", fontsize=14)
plt.xticks([1], ['Direção do Vento'])  # Renomear o eixo x

plt.grid(True)
plt.show()
```
