```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.ticker import MultipleLocator
```

```python
df = pd.read_csv('imigrantes_canada.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
#Tratamento de dados
df.set_index(['País'], inplace=True)
anos = list(map(str,range(1980,2013)))
```

```python
anos
```

```python
brasil = df.loc['Brasil', anos]
brasil
```

```python
argentina = df.loc['Argentina', anos]
argentina
argentina_dic={
    'ano':argentina.index.tolist(),
    'imigrantes': argentina.values.tolist()
}
```

```python
#Criando o dataframe
df_brasil = pd.DataFrame(brasil_dic)
df_brasil
```

```python
df_brasil.info()
```

```python
#Criando o dataframe
df_argentina = pd.DataFrame(argentina_dic)
df_argentina
```

```python
df_argentina.info()
```

```python
plt.plot(df_brasil['ano'], df_brasil['imigrantes'], label = 'brasil')
plt.plot(df_argentina['ano'], df_argentina['imigrantes'], label = 'argentina')
plt.xticks(['1980', '1985', '1990', '1995', '2000', '2005', '2010'])
plt.yticks([500,1000,1500,2000,2500,3000])
plt.xlabel("Anos",fontsize=14)
plt.legend()
plt.ylabel("Número de Imigrantes", fontsize=14)
plt.title('Imigração da Argentina para o Canadá', fontsize=18)
plt.show()

```

```python
df_comparacao = df.loc[['Brasil', 'Argentina'], anos]
df_comparacao = df_comparacao.T
df_comparacao.head()
```

```python
df_comparacao['Total']=df_comparacao['Brasil']+df_comparacao['Argentina']
```

```python
plt.figure(figsize=(10,10))
plt.plot(df_comparacao['Brasil'], label = 'Brasil', color= 'blue')
plt.plot(df_comparacao['Argentina'], label = 'Argentina', color='red')
plt.plot(df_comparacao['Total'], label= 'Total', color='black')
plt.xlabel('Anos', fontsize=14)
plt.ylabel('Número de Imigrantes', fontsize=14)
plt.title("Comparação de imigrantes do Brasil e Argentina para o Canadá")
plt.xticks(['1980', '1985', '1990', '1995', '2000', '2005', '2010'])
plt.yticks([500,1000,1500,2000,2500,3000,3500,4000], fontsize=14)
plt.legend(loc='upper left')
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(8,4))

ax.plot(df_brasil['ano'], df_brasil['imigrantes'], lw=3, color = 'green')
ax.xaxis.set_major_locator(plt.MultipleLocator(5))
ax.xaxis.set_tick_params(labelsize=12)
ax.yaxis.set_tick_params(labelsize=12)
ax.grid(linestyle='--')
ax.set_title('Imigração do Brasil para o Canadá \n 1980 a 2013', loc='left')
ax.set_xlabel('Ano', fontsize=14)
ax.set_ylabel('Nº de imigrantes', fontsize=14)
ax.legend(loc='upper left')
plt.show()
```

```python
fig, axs = plt.subplots(1,2, figsize=(15,5))

axs[0].plot(df_brasil['ano'], df_brasil['imigrantes'], lw=3, color = 'green')
axs[0].set_title('Imigração do Brasil para o Canadá \n 1980 a 2013')
axs[0].set_xlabel('Ano')
axs[0].set_ylabel('Nº de imigrantes')

axs[0].xaxis.set_major_locator(plt.MultipleLocator(5))
axs[0].grid()

#Subplot 2: Boxplot

axs[1].boxplot(df_brasil['imigrantes'])
axs[1].set_title('Boxplot da imigração do Brasil para Canadá \n1980 a 2013')
axs[1].set_xlabel('Brasil')
axs[1].set_ylabel('Número de Imigrantes')
axs[1].grid()

plt.show()
```
