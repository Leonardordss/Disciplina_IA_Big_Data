```python
# Importando as bibliotecas necess√°rias
import numpy as np
import pandas as pd
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
```

```python
#Definindo os dados

dia_semana = [1,2,3,4,5,6,7]
n_de_clientes= [15,20,30,100,350,500,700]
```

```python
# Criando um DataFrame 
df = pd.DataFrame({
    'Dias': dia_semana,
    'Clientes': n_de_clientes
})
```

```python
# Exibindo o DataFrame
print("DataFrame")
print(df)
```

```python
# Aplicando o algoritmo KMeans
kmeans = KMeans(n_clusters=3, random_state=0)
df['Cluster'] = kmeans.fit_predict(df[['Clientes']])
```

```python
# Exibindo o DataFrame com os clusters
print("\nDataFrame com Clusters:")
print(df)
```

```python
# Plotando os resultados
plt.scatter(df['Dias'], df['Clientes'], c=df['Cluster'], cmap='viridis', marker='o')
plt.scatter(kmeans.cluster_centers_, kmeans.predict(kmeans.cluster_centers_.reshape(-1, 1)), marker='x', color='red', s=100, label='Centroides')
plt.xlabel('Dias')
plt.ylabel('Clientes')
plt.title('Agrupamento Usando KMeans')
plt.legend()
plt.show()
```
