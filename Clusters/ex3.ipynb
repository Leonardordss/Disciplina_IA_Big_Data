```python
# Importando as bibliotecas necess√°rias
import numpy as np
import pandas as pd
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
```

```python
# Definindo os dados
setores = np.array([1, 2, 3, 4, 5, 6, 7, 8])
produtos_fabricados = np.array([100, 50, 15, 200, 500, 1000, 375, 450])
```

```python
# Criando um DataFrame 
df = pd.DataFrame({
    'Setor': setores,
    'Produtos Fabricados': produtos_fabricados
})
```

```python
# Exibindo o DataFrame
print("DataFrame")
print(df)
```

```python
# Aplicando o algoritmo KMeans
kmeans = KMeans(n_clusters=3, random_state=0)
df['Cluster'] = kmeans.fit_predict(df[['Produtos Fabricados']])
```

```python
# Exibindo o DataFrame com os clusters
print("\nDataFrame com Clusters:")
print(df)
```

```python
# Plotando os resultados
plt.scatter(df['Setor'], df['Produtos Fabricados'], c=df['Cluster'], cmap='viridis', marker='o')
plt.scatter(kmeans.cluster_centers_, kmeans.predict(kmeans.cluster_centers_.reshape(-1, 1)), marker='x', color='red', s=100, label='Centroides')
plt.xlabel('Setor')
plt.ylabel('Produtos Fabricados')
plt.title('Agrupamento dos Setores Usando KMeans')
plt.legend()
plt.show()
```
